## **PUNTO 2**

```python
class Punto:
  def __init__(self, x, y):
    self.x = int(x)
    self.y = int(y)

  def eje_x(self):
    return self.x

  def eje_y(self):
    return self.y

  def impresion(self):
    return f"({self.x},{self.y})"

  def opuesto (self):
    x = self.x * -1
    y = self.y * -1
    return f"({x},{y})"
```

## **PUNTO 4**

```python
class Cancion:
    def __init__(self, titulo, autor):
        self.titulo = str(titulo)
        self.autor = str(autor)

    def cancion(self):
        return f"La canción reproducida es {self.titulo} de {self.autor}"

    def get_titulo(self):
        return f"El título de la canción es {self.titulo}"

    def get_autor(self):
        return f"El autor es {self.autor}"

    def set_titulo(self, new_title):
        self.titulo = str(new_title)

    def set_autor(self, new_author):
        self.autor = str(new_author)
```

## **PUNTO 5**

```python
class Libro:
  def __init__ (self, titulo, autor, ISBN, paginas, edicion, editorial, lugar, fecha_edicion):
    self.titulo = titulo
    self.autor = autor
    self.ISBN = ISBN
    self.paginas = paginas
    self.edicion = edicion
    self.editorial = editorial
    self.lugar = lugar
    self.fecha_edicion = str(fecha_edicion)

  def lectura(self):
        return (
            f"Titulo:\t{self.titulo}\n"
            f"Autor:\t{self.autor}\n"
            f"ISBN:\t{self.ISBN}\n"
            f"Paginas:\t{self.paginas}\n"
            f"Edicion:\t{self.edicion}\n"
            f"Editorial:\t{self.editorial}\n"
            f"Lugar:\t{self.lugar}\n"
            f"Fecha de edición:\t{self.fecha_edicion}"
        )

  def titulo(self):
    return f"El titulo del libro es {self.titulo}"

  def autor(self):
    return f"El autor del libro es {self.autor}"

  def ISBN(self):
    return f"El ISBN del libro es {self.ISBN}"

  def paginas(self):
    return f"La cantidad de paginas del libro es {self.paginas}"

  def edicion(self):
    return f"La edicion del libro es {self.edicion}"

  def editorial(self):
    return f"La editorial del libro es {self.editorial}"

  def lugar(self):
    return f"El lugar del libro es {self.lugar}"

  def fecha_de_edicion(self):
    return f"La fecha de edicion del libro es {self.fecha_edicion}"
```
